```python
import pandas as pd
import numpy as np


dados_produtos = {'Produto': ['Higiene', 'Churrasqueira', 'Ração Gatos', 'Cadeiras', 'Mesas', 'Arroz', 'Notebook', 'Produto Higiene', 'Churrasqueira', 'Ração Gatos', 'Cadeiras', 'Mesas', 'Arroz', 'Notebook', 'Geladeira']}
df_produtos = pd.DataFrame(dados_produtos)


dados_pedidos = {'Rastreio': ['R001', 'R002', 'R003', 'R004', 'R005', 'R006', 'R007', 'R008', 'R009', 'R010', 'R011', 'R012', 'R013', 'R014', 'R015'],
                 'UF': ['SP', 'RJ', 'MG', 'PR', 'SC', 'RS', 'BA', 'PE', 'CE', 'AM', 'PA', 'GO', 'MT', 'MS', 'TO'],
                 'Data do Pedido': pd.date_range(start='2022-01-01', end='2022-01-15', periods=15),
                 'Cliente': ['João', 'Maria', 'José', 'Ana', 'Pedro', 'Paula', 'Carlos', 'Mariana', 'Lucas', 'Julia', 'Fernando', 'Gabriel', 'Isabela', 'Luana', 'Rafael'],
                 'Prazo de Entrega': np.random.randint(1, 10, size=15),
                 'Descrição do Produto': df_produtos['Produto'].sample(n=15, replace=True).tolist()}

df_pedidos = pd.DataFrame(dados_pedidos)


df_pedidos['Previsão de Entrega'] = df_pedidos['Data do Pedido'] + pd.to_timedelta(df_pedidos['Prazo de Entrega'], unit='D')


df_pedidos['Data da Entrega'] = pd.date_range(start='2022-01-05', end='2022-01-19', periods=15)


df_pedidos['Data da Entrega'] = df_pedidos['Data da Entrega'].dt.date


df_pedidos['Status do Pedido'] = np.where(df_pedidos['Data da Entrega'] <= df_pedidos['Previsão de Entrega'], 'No Prazo', 'Atrasado')


df_pedidos['Etapa do Pedido'] = np.where(df_pedidos['Data da Entrega'].isnull(), 'Em Andamento', 'Pedido Entregue')

df_pedidos.head(15)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

sla_dentro_prazo = (df_pedidos['Status do Pedido'] == 'No Prazo').sum() / len(df_pedidos) * 100
sla_atrasado = (df_pedidos['Status do Pedido'] == 'Atrasado').sum() / len(df_pedidos) * 100

labels = ['Dentro do Prazo', 'Atrasado']
sizes = [sla_dentro_prazo, sla_atrasado]
colors = ['gray', 'blue']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('SLA de Entrega')
plt.show()
```

```python
df_uf = df_pedidos.groupby('UF')['Rastreio'].count().reset_index()
mapa = folium.Map(location=[-15.788497,-47.879873], zoom_start=4)
for i, row in df_uf.iterrows():
    uf = row['UF']
    count = row['Rastreio']
    folium.Marker(location=get_coordinates(uf), popup=f'{uf}: {count} pedidos').add_to(mapa)
mapa

```
